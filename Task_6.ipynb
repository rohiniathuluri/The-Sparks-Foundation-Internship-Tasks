```python
import pandas as pd
import sklearn.datasets as datasets
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.isna().any()
```

```python
df.info()
```

```python
y=iris.target
```

```python
y
```

```python
df.describe()
```

```python
df.shape
```

```python
import seaborn as sns
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,Y_train,y_test = train_test_split(df,y,test_size=0.2,random_state=6) 
```

```python
classifier = DecisionTreeClassifier()
```

```python
classifier.fit(X_train,Y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
fig = plt.figure(figsize = (16,15))
```

```python
plot_tree(classifier,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,rounded=True)
plt.show()
```

```python
#Finding accuracy and scores
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error,r2_score
```

```python
y_preds = classifier.predict(X_train)
```

```python
y_preds
```

```python
Y_train
```

```python
accuracy_1 = metrics.accuracy_score(y_preds,Y_train)
```

```python
y_preds_2 = classifier.predict(x_test)
```

```python
y_preds_2
```

```python
y_test
```

```python
accuracy_2 = metrics.accuracy_score(y_test,y_preds_2)
```

```python
print("The accuracies of training and testing models are: ")
print("The accuracy of training model is:",accuracy_1)
print("The accuracy of testing model is:",accuracy_2)
print("Done!!--..")
```
